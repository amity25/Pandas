```python
#1. Import Pandas
import pandas as pd
```

```python
'''2. Create Sample Dirty Dataset

We will first create a dataset with:

Missing values
Duplicate rows
Wrong data types'''

data = {
    "name": ["Amit", "Rahul", "Sneha", "Amit", None],
    "sales": [500, 700, None, 500, 900],
    "city": ["Delhi", "Mumbai", "Delhi", "Delhi", None]
}

df = pd.DataFrame(data)

print(df)
```

```python
'''Understanding the Problems in This Dataset
Problem 1 — Missing Values

Some cells are empty:

None
NaN
Problem 2 — Duplicate Rows

Row 0 and Row 3 are duplicates.

Problem 3 — Wrong Data Types

Sales should ideally be integers.'''
```

```python
'''PART 1 — Missing Values
What are Missing Values?

Missing values are empty cells in a dataset.

Examples:

Blank Excel cells
Null database values
Missing customer information

Pandas usually shows missing values as:

NaN → Not a Number
None'''
```

```python
'''Check Missing Values
df.isnull()

This checks every cell and returns:

True → value is missing
False → value exists'''
```

```python
print(df.isnull())
```

```python
#Count Missing Values in Each Column
print(df.isnull().sum())
'''Explanation
isnull() finds missing values
sum() counts True values

Since:

True = 1
False = 0

Pandas counts total missing values.'''
```

```python
'''Remove Missing Values
df.dropna()

This removes rows containing missing values.'''
clean_df = df.dropna()

print(clean_df)

'''Explanation

Rows with:

Missing sales
Missing name
Missing city

were removed.'''
```

```python
'''Important Note

dropna() can remove a lot of data.

In real projects:

Sometimes removing rows is okay
Sometimes we fill missing values instead'''
```

```python
'''Fill Missing Values
df.fillna()

Instead of deleting rows, we can fill empty cells.'''

#Example 1 — Fill with 0
df["sales"] = df["sales"].fillna(0)

print(df)
```

```python
#Example 2 — Fill Text Columns
df["city"] = df["city"].fillna("Unknown")

print(df)
```

```python
'''Common fillna() Methods
Method	Purpose
fillna(0)	Replace missing numbers
fillna("Unknown")	Replace missing text
fillna(df.mean())	Replace with average
fillna(method="ffill")	Fill using previous row'''
```

```python
'''PART 2 — Remove Duplicates
What are Duplicate Rows?

Duplicate rows are repeated records.

Example:

Same customer entered twice
Same sales record repeated
System importing duplicate data

Duplicates can:

Increase totals incorrectly
Create wrong reports
Affect business decisions'''
```

```python
#Check Duplicate Rows
print(df.duplicated())
'''Explanation

Row 3 is duplicate because:

It matches row 0'''
```

```python
#Remove Duplicate Rows
df.drop_duplicates()
df = df.drop_duplicates()

print(df)
```

```python
'''Important Parameters
Keep First Duplicate (Default)'''
df.drop_duplicates(keep="first")
```

```python
#Keep Last Duplicate
df.drop_duplicates(keep="last")
```

```python
#Remove ALL Duplicates
df.drop_duplicates(keep=False)
```

```python
'''PART 3 — Rename Columns
Why Rename Columns?

Real datasets often have:

Bad column names
Spaces
Short names
Inconsistent formatting

Example:

sales
Sales Amount
TOTALSALE
sales_amt

We rename columns for:

Better readability
Easier coding
Standard naming'''
```

```python
#Rename Single Column
#df.rename()
df = df.rename(columns={"sales": "total_sales"})

print(df)
```

```python
#Rename Multiple Columns
df = df.rename(columns={
    "name": "customer_name",
    "city": "customer_city"
})

print(df)
```

```python
'''Best Practices for Column Names

Use:

lowercase
underscore _
meaningful names

Good:

total_sales
customer_name
order_date

Avoid:

Total Sales
TOTALSALES
x1'''
```

```python
'''PART 4 — Change Data Types
Why Data Types Matter

Sometimes numbers are stored as:

Text (string)

Example:

"500"

instead of:

500

This creates problems in:

Calculations
Sorting
Charts
Analysis'''
```

```python
#Check Data Types
print(df.dtypes)
```

```python
'''Common Pandas Data Types
Data Type	Meaning
int64	Integer
float64	Decimal
object	Text/String
bool	True/False
datetime64	Date & Time'''
```

```python
#Convert Data Type
#astype()
df["total_sales"] = df["total_sales"].astype(int)

print(df.dtypes)
```

```python
'''Important Warning

If a column contains:

Text
Missing values
Invalid numbers

Then astype(int) may produce errors.

Example:

"abc"

cannot become integer.'''
```

```python
#FULL DATA CLEANING EXAMPLE
import pandas as pd

data = {
    "name": ["Amit", "Rahul", "Sneha", "Amit", None],
    "sales": [500, 700, None, 500, 900],
    "city": ["Delhi", "Mumbai", "Delhi", "Delhi", None]
}

df = pd.DataFrame(data)

# Check missing values
print(df.isnull().sum())

# Fill missing sales with 0
df["sales"] = df["sales"].fillna(0)

# Fill missing city
df["city"] = df["city"].fillna("Unknown")

# Remove duplicates
df = df.drop_duplicates()

# Rename column
df = df.rename(columns={"sales": "total_sales"})

# Change data type
df["total_sales"] = df["total_sales"].astype(int)

print(df)

print(df.dtypes)
```
