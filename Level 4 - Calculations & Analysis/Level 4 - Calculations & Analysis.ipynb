```python
#2. Setup the Example Dataset
import pandas as pd

data = {
    "order_id": [1, 2, 3, 4],
    "region": ["North", "South", "North", "West"],
    "sales": [100, 200, 150, 300]
}

df = pd.DataFrame(data)

print(df)
```

```python
'''3. SUM — Total Value
Purpose

Adds all values together.

Syntax
df["sales"].sum()'''
total_sales = df["sales"].sum()

print(total_sales)
```

```python
'''4. MEAN — Average
Purpose

Finds the average value.

Syntax
df["sales"].mean()'''

average_sales = df["sales"].mean()

print(average_sales)
```

```python
'''Explanation

Formula:

Average=
4
100+200+150+300
	​
'''
```

```python
'''5. MAX — Highest Value
Purpose

Finds the largest value.

Syntax
df["sales"].max()'''

highest_sale = df["sales"].max()

print(highest_sale)
```

```python
'''MIN — Lowest Value'''
df["sales"].min()
```

```python
'''6. COUNT — Number of Records'''
df["sales"].count()
```

```python
'''Multiple Calculations Together'''
print("Total:", df["sales"].sum())
print("Average:", df["sales"].mean())
print("Maximum:", df["sales"].max())
print("Minimum:", df["sales"].min())
```

```python
'''What is Group By?

It means:

“Group similar data together, then perform calculations.”'''

region_sales = df.groupby("region")["sales"].sum()

print(region_sales)
```

```python
'''9. Group By with Mean

Instead of total sales:'''

df.groupby("region")["sales"].mean()
```

```python
'''10. Group By with Multiple Calculations'''
df.groupby("region")["sales"].agg(["sum", "mean", "max"])

```

```python
'''11. describe() — Quick Full Analysis

This is a very important shortcut.'''

df["sales"].describe()
```
