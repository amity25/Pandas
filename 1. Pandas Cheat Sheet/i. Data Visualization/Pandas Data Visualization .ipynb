```python
'''Step 1: Import Libraries
Display charts inside Jupyter Notebook
%matplotlib inline
'''

import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''Step 2: Create Sample Dataset

We'll use the same dataset throughout this lesson.'''

import pandas as pd

sales_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [12000, 15000, 18000, 17000, 22000, 25000],
    "Expenses": [7000, 8000, 9000, 8500, 11000, 12000],
    "Customers": [120, 135, 150, 145, 170, 190]
}

df = pd.DataFrame(sales_data)

df
```

```python
'''1. Line Plot

Used to visualize trends over time.

Use Cases
Sales growth
Stock prices
Website traffic
Revenue trends'''

df.plot(
    x="Month",
    y="Sales",
    kind="line",
    title="Monthly Sales Trend",
    figsize=(8,5)
)

plt.show()
```

```python
'''2. Vertical Bar Plot

Compares values across categories.
Use Cases
Sales by month
Employees by department
Products by revenue'''

df.plot(
    x="Month",
    y="Sales",
    kind="bar",
    title="Monthly Sales",
    figsize=(8,5)
)

plt.show()
```

```python
'''3. Horizontal Bar Plot

Useful when category names are long.
Use Cases
Ranking charts
Survey results
Performance comparisons'''

df.plot(
    x="Month",
    y="Sales",
    kind="barh",
    title="Monthly Sales",
    figsize=(8,5)
)

plt.show()
```

```python
'''4. Histogram

Shows data distribution.

What It Shows
Frequency of values
Data spread
Skewness
Use Cases
Salary analysis
Exam scores
Product prices'''

df["Sales"].plot(
    kind="hist",
    bins=5,
    title="Sales Distribution",
    figsize=(8,5)
)

plt.show()
```

```python
'''5. Box Plot

Displays statistical summary.

Shows

Median
Quartiles
Outliers
Data spread

Use Cases

Detect unusual values
Compare distributions
Data quality checks'''

df[["Sales", "Expenses"]].plot(
    kind="box",
    figsize=(8,5),
    title="Box Plot"
)

plt.show()
```

```python
'''6. KDE Plot (Kernel Density Estimation)

A smooth version of a histogram.

What It Shows
The probability distribution of data.

Use Cases
Statistical analysis
Distribution comparison
Trend identification'''

df["Sales"].plot(
    kind="kde",
    figsize=(8,5),
    title="Sales Density Curve"
)

plt.show()
```

```python
'''7. Pie Chart

Shows percentage contribution.

Use Cases

Market share
Budget allocation
Revenue contribution'''

df.set_index("Month")["Sales"].plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(7,7),
    title="Sales Share by Month"
)

plt.ylabel("")
plt.show()
```

```python
'''8. Scatter Plot

Shows relationship between two variables.


What It Shows

Whether two variables move together.

Use Cases
Sales vs Customers
Height vs Weight
Ad Spend vs Revenue'''

df.plot.scatter(
    x="Customers",
    y="Sales",
    figsize=(8,5),
    title="Customers vs Sales"
)

plt.show()
```

```python
'''9. Area Plot

Shows cumulative trends.

Use Cases
Revenue growth
Resource utilization
Market trends'''

df.plot(
    x="Month",
    y=["Sales", "Expenses"],
    kind="area",
    figsize=(8,5),
    title="Sales and Expenses Area Chart"
)

plt.show()
```

```python
'''Multiple Lines in One Chart

Compare multiple columns together.'''

df.plot(
    x="Month",
    y=["Sales", "Expenses"],
    kind="line",
    figsize=(8,5),
    title="Sales vs Expenses"
)

plt.show()
```

```python
'''Customizing Charts

Add labels and titles.'''

df.plot(
    x="Month",
    y="Sales",
    kind="line",
    figsize=(8,5)
)

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.grid(True)

plt.show()
```

```python
'''Save Chart as Image'''

df.plot(
    x="Month",
    y="Sales",
    kind="line"
)

plt.savefig("sales_chart.png")
plt.show()
```
