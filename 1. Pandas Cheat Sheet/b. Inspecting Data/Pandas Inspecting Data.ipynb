```python
#Step 1 — Import Pandas
import pandas as pd
```

```python
'''Step 2 — Create or Load a DataFrame

Example DataFrame:'''

data = {
    "Name": ["Amit", "Sara", "John", "Priya", "David"],
    "Age": [28, 24, 32, 29, 35],
    "Salary": [50000, 45000, 70000, 52000, 80000],
    "Department": ["IT", "HR", "Finance", "IT", "Management"]
}

df = pd.DataFrame(data)

print(df)
```

```python
'''1. df.head() — View First Rows

This command displays the first 5 rows of the DataFrame'''

df.head()
```

```python
'''Custom Number of Rows

You can specify how many rows you want.'''

df.head(2)
```

```python
'''2. df.tail() — View Last Rows

Shows the last 5 rows.'''

df.tail()
```

```python
df.tail(2)
```

```python
'''3. df.sample() — Random Rows

Displays random rows from the dataset.'''

df.sample()
```

```python
'''Why is sample() Important?

Very useful when working with huge datasets.

Instead of checking only top rows, random sampling helps you:

Inspect random records
Detect inconsistent values
Validate data quality
Perform quick spot-checks'''
```

```python
'''Multiple Random Rows
Output changes every time.'''

df.sample(3)
```

```python
'''4. df.shape — Dataset Dimensions

Returns number of rows and columns.'''

import pandas as pd

data = {
    "Name": ["Amit", "Sara", "John", "Priya", "David"],
    "Age": [28, 24, 32, 29, 35],
    "Salary": [50000, 45000, 70000, 52000, 80000],
    "Department": ["IT", "HR", "Finance", "IT", "Management"]
}

df = pd.DataFrame(data)

df.shape
```

```python
'''Access Individually
Number of Rows'''

import pandas as pd

data = {
"Name": ["Amit", "Sara", "John", "Priya", "David"],
    "Age": [28, 24, 32, 29, 35],
    "Salary": [50000, 45000, 70000, 52000, 80000],
    "Department": ["IT", "HR", "Finance", "IT", "Management"]
}

df = pd.DataFrame(data)

df.shape[0]
```

```python
'''Number of Columns'''

import pandas as pd

data = {
    "Name": ["Amit", "Sara", "John"],
    "Age": [28, 24, 32]
}

df = pd.DataFrame(data)

df.shape[1]
```

```python
'''5. df.info() — Complete Dataset Summary

One of the MOST IMPORTANT inspection commands.

What Does info() Tell You?
Information	Meaning
Total rows	Number of entries
Column names	Names of all columns
Non-null count	Missing value detection
Data types	int, float, object, etc.
Memory usage	RAM usage

Common Data Types in Pandas
Data Type	Meaning
int64	Integer numbers
float64	Decimal numbers
object	Text/String
bool	True/False
datetime64	Dates and time'''

df.info()
```

```python
'''6. df.describe() — Statistical Summary

Generates statistics for numerical columns.

Meaning of Each Statistic
Statistic	Meaning
count	Number of values
mean	Average
std	Standard deviation
min	Smallest value
25%	First quartile
50%	Median
75%	Third quartile
max	Largest value'''

df.describe()
```

```python
'''Include All Columns

By default, only numerical columns are shown.

To include text columns:'''

df.describe(include='all')
```

```python
'''7. df.dtypes — Check Column Data Types'''

df.dtypes
```

```python
'''8. df.columns — View Column Names'''

df.columns
```
