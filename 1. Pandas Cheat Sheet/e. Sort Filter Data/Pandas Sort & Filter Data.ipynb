```python
'''. Sort Data in Ascending Order

To sort rows based on a column in ascending order:'''

import pandas as pd

data = {
    "Name": ["Amit", "Riya", "Karan", "Neha"],
    "Marks": [78, 92, 65, 88]
}

df = pd.DataFrame(data)

sorted_df = df.sort_values('Marks')

print(sorted_df)
```

```python
'''2. Sort Data in Descending Order

To sort from highest to lowest:'''

sorted_df = df.sort_values('Marks', ascending=False)

print(sorted_df)
```

```python
'''3. Sort by Multiple Columns

You can sort using more than one column.'''

data = {
    "Name": ["Amit", "Riya", "Karan", "Neha"],
    "Department": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 70000, 60000, 65000]
}

df = pd.DataFrame(data)

sorted_df = df.sort_values(
    ['Department', 'Salary'],
    ascending=[True, False]
)

print(sorted_df)
```

```python
'''4. Filter Rows Using Conditions

Filtering means selecting only rows that match a condition.'''

data = {
    "Name": ["Amit", "Riya", "Karan", "Neha"],
    "Marks": [78, 92, 65, 88]
}

df = pd.DataFrame(data)

filtered_df = df[df['Marks'] > 80]

print(filtered_df)
```

```python
'''5. Filter Using query()

Pandas provides a cleaner filtering method using SQL-like syntax.'''

filtered_df = df.query('Marks > 80')

print(filtered_df)
```

```python
'''6. Select Random Rows with sample()

Useful for testing, previews, or random analysis.'''

sample_df = df.sample(2)

print(sample_df)
```

```python
'''7. Get Largest Values with nlargest()

To get top rows based on highest values:'''

top_df = df.nlargest(2, 'Marks')

print(top_df)
```

```python
'''8. Get Smallest Values with nsmallest()

To get lowest values:'''

bottom_df = df.nsmallest(2, 'Marks')

print(bottom_df)
```

```python
'''9. Filter Columns Using filter()

This filters column names based on text patterns.'''

data = {
    "emp_name": ["Amit", "Riya"],
    "emp_salary": [50000, 70000],
    "dept": ["IT", "HR"]
}

df = pd.DataFrame(data)

filtered_columns = df.filter(like='emp')

print(filtered_columns)
```

```python
'''Key Points to Remember
Function	Purpose
sort_values()	Sort rows
ascending=False	Descending sorting
query()	SQL-like filtering
sample()	Random rows
nlargest()	Top values
nsmallest()	Lowest values
filter()	Filter column names'''
```
