```python
import pandas as pd
```

```python
#Sample Dataset

data = {
    'Name': ['Amit', 'Rahul', None, 'Sneha', 'Amit'],
    'Age': [25, None, 30, 28, 25],
    'Salary': [50000, 60000, None, 75000, 50000]
}

df = pd.DataFrame(data)

print(df)
```

```python
'''🔹 1. df.isnull()
Check Missing Values

This function checks where null (missing) values exist.'''

df.isnull()
```

```python
#🔹 Count Missing Values

df.isnull().sum()
```

```python
'''🔹 2. df.notnull()
Check Non-Missing Values

Opposite of isnull().'''

df.notnull()
```

```python
'''🔹 3. df.dropna()

Remove Missing Values

Removes rows containing null values.'''

clean_df = df.dropna()

print(clean_df)
```

```python
'''🔹 Remove Columns with Missing Values
df.dropna(axis=1)
Explanation
axis=0 → Remove rows
axis=1 → Remove columns'''
```

```python
'''🔹 4. df.fillna()
Replace Missing Values

Instead of deleting data, missing values can be filled.'''

df.fillna(0)
```

```python
#Fill Specific Column

df['Age'] = df['Age'].fillna(18)

print(df)

#Only missing values in the Age column are replaced.
```

```python
'''🔹 5. df.replace()
Replace Specific Values

Used to replace existing values.

Replace a Single Value'''

df.replace(25, 'Twenty Five')
```

```python
#Replace Multiple Values

df.replace({
    25: 'Young',
    30: 'Adult'
})
```

```python
'''🔹 6. df.rename()
Rename Columns

Useful when column names are unclear or inconsistent.'''

df.rename(columns={'Salary': 'Monthly_Salary'})
```

```python
#🔹 Rename Multiple Columns

df.rename(columns={
    'Name': 'Employee_Name',
    'Age': 'Employee_Age'
})
```

```python
'''🔹 7. df.astype()
Change Data Types

Used to convert one data type into another.'''

df['Age'] = df['Age'].fillna(0).astype('int')

print(df.dtypes)
```

```python
'''🔹 Common Data Types
Data Type	Meaning
int	Integer
float	Decimal Number
object	Text/String
bool	True/False'''
```

```python
'''🔹 8. df.drop_duplicates()
Remove Duplicate Rows

Duplicate data is very common in real datasets.'''

df.drop_duplicates()
```

```python
'''🔹 Remove Duplicates Based on One Column'''

df.drop_duplicates(subset=['Name'])
```

```python
'''🔹 9. df.reset_index()
Reset Index Numbers

After removing rows, indexes may become irregular.'''

new_df = df.dropna()

print(new_df)
```

```python
'''Reset the Index'''

new_df.reset_index()
```

```python
#Remove Old Index Column

new_df.reset_index(drop=True)
```

```python
#🔹 Real-World Cleaning Example

df = pd.read_csv(r"C:\Users\amita\Desktop\employees.csv")

# Remove duplicates
df = df.drop_duplicates()

# Fill missing salaries
df['Salary'] = df['Salary'].fillna(0)

# Rename columns
df = df.rename(columns={'EmpName': 'Employee_Name'})

# Reset index
df = df.reset_index(drop=True)

print(df.head())
```
