```python
'''1. Create a Sample DataFrame'''

import pandas as pd

data = {
    'Name': ['Amit', 'Riya', 'John', 'Sara', 'David'],
    'Age': [28, 24, 35, 29, 40],
    'Salary': [50000, 45000, 70000, 52000, 90000],
    'Department': ['IT', 'HR', 'Finance', 'IT', 'Management']
}

df = pd.DataFrame(data)

print(df)
```

```python
'''2. Select a Single Column
Syntax
df['column']'''

print(df['Name'])
```

```python
'''3. Select Multiple Columns
Syntax
df[['col1', 'col2']]'''

print(df[['Name', 'Salary']])
```

```python
'''4. Select Rows Using iloc
Syntax
df.iloc[row_position]'''

print(df.iloc[0])
```

```python
'''5. Select Rows Using loc
Syntax
df.loc[row_label]'''

print(df.loc[0])
```

```python
'''6. Select a Specific Element Using iloc
Syntax
df.iloc[row_position, column_position]'''

print(df.iloc[0, 0])
```

```python
'''7. Select a Specific Element Using loc
Syntax
df.loc[row_label, 'column']'''

print(df.loc[0, 'Salary'])
```

```python
'''8. Filter Rows with Conditions
Syntax
df[df['column'] > value]'''

print(df[df['Salary'] > 50000])
```

```python
'''9. Slice Rows and Columns
Syntax
df.iloc[row_start:row_end, col_start:col_end]
Example'''

print(df.iloc[0:3, 0:2])
```

```python
'''10. Set a Column as Index
Syntax
df.set_index('column')'''

new_df = df.set_index('Name')

print(new_df)
```

```python
'''Difference Between loc and iloc

| Feature | loc | iloc |
|---|---|
| Works With | Labels | Positions |
| Input Type | Names | Integer Numbers |
| Example | df.loc[0, 'Name'] | df.iloc[0, 0] |'''
```
