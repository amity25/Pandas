```python
import pandas as pd
```

```python
# Create Sample DataFrame

data = {
    'Department': ['HR', 'IT', 'HR', 'Finance', 'IT', 'Finance', 'HR'],
    'Employee': ['Amit', 'Rahul', 'Priya', 'Neha', 'Karan', 'Simran', 'Rohit'],
    'Salary': [40000, 70000, 45000, 60000, 80000, 65000, 50000],
    'Bonus': [5000, 7000, 4000, 6000, 8000, 6500, 4500],
    'Experience': [2, 5, 3, 4, 6, 5, 2]
}

df = pd.DataFrame(data)

print(df)
```

```python
# 1. Group By a Column
#It simply means the groups were created successfully

grouped = df.groupby('Department')

print(grouped)
```

```python
#To Actually View the Groups

#Option 1 — Loop Through Groups

grouped = df.groupby('Department')

for department, data in grouped:
    print("\nDepartment:", department)
    print(data)
```

```python
# Option 2 — View a Specific Group

grouped = df.groupby('Department')

print(grouped.get_group('HR'))
```

```python
# Option 3 — Count Rows in Each Group

grouped = df.groupby('Department')

print(grouped.size())
```

```python
#Option 4 — Apply Aggregation Directly
#This is the most commonly used approach in real projects.

print(df.groupby('Department').mean(numeric_only=True))

```

```python
# 2. Mean of Groups


mean_values = df.groupby('Department').mean(numeric_only=True)

print(mean_values)
```

```python
# 3. Sum of Groups

sum_values = df.groupby('Department').sum(numeric_only=True)

print(sum_values)
```

```python
# 4. Count Non-Null Values in Groups

count_values = df.groupby('Department').count()

print(count_values)
```

```python
# 5. Maximum Salary in Each Department

max_salary = df.groupby('Department')['Salary'].max()

print(max_salary)
```

```python
# 6. Create Pivot Table
pivot_table = df.pivot_table(
    values='Salary',
    index='Department',
    aggfunc='mean'
)

print(pivot_table)
```

```python
# 7. Aggregate Multiple Columns

aggregated_values = df.agg({
    'Salary': 'mean',
    'Bonus': 'sum'
})

print(aggregated_values)
```

```python
# 8. Apply Function to Columns

column_means = df[['Salary', 'Bonus', 'Experience']].apply('mean')

print(column_means)
```

```python
# 9. Transform Data Column-Wise

transformed_data = df[['Salary', 'Bonus']].transform(
    lambda x: x + 10
)

print(transformed_data)
```

```python
# 10. Multiple Aggregations Example

multiple_agg = df.groupby('Department').agg({
    'Salary': ['mean', 'max', 'min'],
    'Bonus': ['sum', 'mean']
})

print(multiple_agg)
```
