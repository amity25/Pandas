```python
'''1. Concatenate DataFrames Vertically

Use pd.concat() to stack DataFrames row by row.'''

import pandas as pd

df1 = pd.DataFrame({
    'Name': ['Amit', 'Rahul'],
    'Marks': [85, 90]
})

df2 = pd.DataFrame({
    'Name': ['Neha', 'Priya'],
    'Marks': [88, 92]
})

result = pd.concat([df1, df2])

print(result)
```

```python
'''2. Concatenate DataFrames Horizontally

Use axis=1 to combine DataFrames column by column.'''

df1 = pd.DataFrame({
    'Name': ['Amit', 'Rahul']
})

df2 = pd.DataFrame({
    'Marks': [85, 90]
})

result = pd.concat([df1, df2], axis=1)

print(result)
```

```python
'''3. Merge Two DataFrames on a Key

Use merge() to combine DataFrames using a common column.'''

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Amit', 'Rahul', 'Neha']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Marks': [85, 90, 88]
})

result = df1.merge(df2, on='ID')

print(result)
```

```python
'''4. Join DataFrames

Use join() for SQL-style joining based on index values.'''

df1 = pd.DataFrame({
    'Name': ['Amit', 'Rahul']
}, index=[1, 2])

df2 = pd.DataFrame({
    'Marks': [85, 90]
}, index=[1, 2])

result = df1.join(df2)

print(result)
```

```python
'''5. Append Rows to Another DataFrame

append() was deprecated and then completely removed from Pandas because pd.concat() performs the same task more efficiently.'''

import pandas as pd

df1 = pd.DataFrame({
    'Name': ['Amit'],
    'Marks': [85]
})

df2 = pd.DataFrame({
    'Name': ['Rahul'],
    'Marks': [90]
})

result = pd.concat([df1, df2])

print(result)
```

```python
'''6. Outer Join

An outer join keeps all records from both DataFrames.'''

import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Amit', 'Rahul', 'Neha']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Marks': [85, 90, 88]
})

result = pd.merge(df1, df2, how='inner', on='ID')

print(result)
```

```python
'''7. Inner Join

An inner join keeps only matching records.'''

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Amit', 'Rahul', 'Neha']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Marks': [85, 90, 88]
})

result = pd.merge(df1, df2, how='inner', on='ID')

print(result)
```

```python
'''8. Left Join

A left join keeps all records from the left DataFrame.'''

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Amit', 'Rahul', 'Neha']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Marks': [85, 90, 88]
})

result = pd.merge(df1, df2, how='left', on='ID')

print(result)
```

```python
'''9. Right Join

A right join keeps all records from the right DataFrame.'''

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Amit', 'Rahul', 'Neha']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Marks': [85, 90, 88]
})

result = pd.merge(df1, df2, how='right', on='ID')

print(result)
```
