```python
'''Step 1: Import Pandas and Create Sample Data'''

import pandas as pd

# Create sample employee dataset
data = {
    "Employee": ["John", "Sarah", "Mike", "Emma", "David"],
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df)
```

```python
'''1. df.mean() – Column-wise Mean

The mean is the average value of each numeric column.'''

import pandas as pd

data = {
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df.mean())
```

```python
'''2. df.median() – Column-wise Median

Median is the middle value after sorting the data.

Explanation

Sorted salaries:

45000, 50000, 55000, 65000, 75000

Middle value = 55000

Therefore median salary = 55000'''

import pandas as pd

data = {
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df.median())
```

```python
'''3. df.std() – Standard Deviation

Standard deviation measures how spread out the values are from the mean.'''

import pandas as pd

data = {
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df.std())
```

```python
'''4. df.var() – Variance

Variance is the square of the standard deviation.'''

import pandas as pd

data = {
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df.var())
```

```python
'''5. df.sum() – Column-wise Sum'''

import pandas as pd

data = {
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df.sum())
```

```python
'''6. df.min() – Minimum Value

Finds the smallest value in each column.'''


import pandas as pd

data = {
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df.min())
```

```python
'''7. df.max() – Maximum Value

Finds the largest value in each column.'''

import pandas as pd

data = {
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df.max())
```

```python
'''8. df.count() – Count Non-Null Values

Counts non-empty values in each column.'''

import pandas as pd

data = {
    "Age": [25, 30, None, 28, 40],
    "Salary": [45000, 55000, 65000, None, 75000],
    "Experience": [2, 5, 8, 3, None]
}

df = pd.DataFrame(data)

print(df)
print("\nCount of non-null values:")
print(df.count())
```

```python
'''9. df.corr() – Correlation Matrix

Correlation measures the relationship between numeric columns.

Correlation Range
+1  = Perfect Positive Correlation
 0  = No Correlation
-1  = Perfect Negative Correlation'''


import pandas as pd

data = {
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df.corr())
```

```python
'''Bonus: All Statistics at Once with describe()

Instead of running multiple commands, Pandas can generate a complete statistical summary.'''

import pandas as pd

data = {
    "Age": [25, 30, 35, 28, 40],
    "Salary": [45000, 55000, 65000, 50000, 75000],
    "Experience": [2, 5, 8, 3, 12]
}

df = pd.DataFrame(data)

print(df.describe())
```

```python
'''Quick Cheat Sheet

Function	Purpose
df.mean()	Average values
df.median()	Middle value
df.std()	Standard deviation
df.var()	Variance
df.sum()	Total values
df.min()	Smallest value
df.max()	Largest value
df.count()	Count non-null values
df.corr()	Relationship between columns
df.describe()	Complete statistical summary'''
```
