```python
'''pd.DataFrame() – Create DataFrame from Dictionary'''

data = {
    "Name": ["Amit", "Sara", "John"],
    "Age": [28, 31, 35],
    "Department": ["IT", "HR", "Finance"]
}

df = pd.DataFrame(data)

print(df)
```

```python
'''Exporting Data'''
```

```python
'''df.to_csv() – Export to CSV'''
df.to_csv("output.csv", index=False)
```

```python
'''df.to_excel() – Export to Excel'''
df.to_excel("output.xlsx", index=False)
```

```python
'''df.to_sql() – Export to SQL Table
replace → overwrite table
append → add new rows
fail → raise error if table exists'''

import sqlite3

conn = sqlite3.connect("company.db")

df.to_sql("employees", conn, if_exists="replace", index=False)
```

```python
'''df.to_json() – Export to JSON'''
df.to_json("employeees.json", orient="records")
```
