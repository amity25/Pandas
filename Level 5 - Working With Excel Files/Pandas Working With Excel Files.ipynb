```python
#Example 6: Read Every Sheet Separately
import pandas as pd

file_path = r"C:\Users\amita\Desktop\sales.xlsx"

all_sheets = pd.read_excel(
    file_path,
    sheet_name=None
)

for name, df in all_sheets.items():
    print(f"\nSheet Name: {name}")
    print(df.head())
```

```python
# Example 7: Combine All Sheets into One DataFrame

#Very common in business reporting.

import pandas as pd

file_path = r"C:\Users\amita\Desktop\sales.xlsx"

all_sheets = pd.read_excel(
    file_path,
    sheet_name=None
)

combined_df = pd.concat(
    all_sheets.values(),
    ignore_index=True
)

print(combined_df.head())

print("\nTotal Rows:")
print(len(combined_df))
```

```python
#Example 9: Read Only Selected Columns From All Sheets
import pandas as pd

file_path = r"C:\Users\amita\Desktop\sales.xlsx"

all_sheets = pd.read_excel(
    file_path,
    sheet_name=None,
    usecols=["Customer", "Sales"]
)

for sheet_name, df in all_sheets.items():
    print(f"\n{sheet_name}")
    print(df.head())
```

```python
#Example 10: Real-World Business Example

#Read all monthly sheets, combine them, and calculate total sales.

import pandas as pd

file_path = r"C:\Users\amita\Desktop\sales.xlsx"

all_sheets = pd.read_excel(
    file_path,
    sheet_name=None
)

combined_df = pd.concat(
    all_sheets.values(),
    ignore_index=True
)

total_sales = combined_df["Sales"].sum()

print("Total Sales:", total_sales)

combined_df.to_excel(
    r"C:\Users\amita\Desktop\annual_sales.xlsx",
    index=False
)

print("Annual report created successfully!")
```

```python
#Export Selected Data

import pandas as pd

# Read Excel file
file_path = r"C:\Users\amita\Desktop\sales.xlsx"

df = pd.read_excel(file_path)

# Check column names
print("Columns in file:")
print(df.columns)

# Filter rows where Sales > 500
high_sales = df[df["Sales"] > 500]

# Save result
output_file = r"C:\Users\amita\Desktop\high_sales.xlsx"

high_sales.to_excel(output_file, index=False)

print("File saved successfully!")
```
