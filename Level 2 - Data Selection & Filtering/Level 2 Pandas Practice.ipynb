```python
#STEP 1 — Import Pandas
import pandas as pd
```

```python
#STEP 2 — Create Sample Dataset
#Use this dataset throughout the level.

data = {
    "name": ["Amit", "Rahul", "Priya", "Sneha", "John"],
    "sales": [1200, 800, 1500, 700, 2000],
    "region": ["North", "South", "North", "East", "West"]
}

df = pd.DataFrame(data)

print(df)
```

```python
'''PART 1 — Selecting Columns
Sometimes you only need one column from the dataset.
Select Single Column'''
df["sales"]
```

```python
#Select name Column
df["name"]
```

```python
'''PART 2 — Selecting Multiple Columns
You can select more than one column together.
Syntax'''
df[["name", "sales"]]
```

```python
#Select 3 Columns
df[["name", "sales", "region"]]
```

```python
#IMPORTANT NOTE
#Why Double Brackets?

#Single Bracket
df["sales"]
#Returns a Series


#Double Bracket
df[["sales"]]
#Returns a DataFrame
```

```python
'''PART 3 — Filtering Data
Filtering means showing only rows that match a condition.
This is one of the MOST important skills in data analysis.
Filter Sales Greater Than 1000'''
df[df["sales"] > 1000]
```

```python
#Understanding the Syntax
#Inside Condition

df["sales"] > 1000

#This creates True/False values.
#Then pandas keeps only True rows.
```

```python
#PART 4 — Filtering Using Conditions
#Filter Region = North
df[df["region"] == "North"]
```

```python
'''Equality Operator
Double Equal Sign
==

Means:
Check if equal to'''
```

```python
''''''Common Comparison Operators
Operator	Meaning
>	Greater than
<	Less than
>=	Greater than equal to
<=	Less than equal to
==	Equal to
!=	Not equal to'''
```

```python
'''PART 5 — Multiple Conditions
Sales Greater Than 1000 AND Region North'''
df[(df["sales"] > 1000) & (df["region"] == "North")]
```

```python
'''Important Rules
Use & for AND
(condition1) & (condition2)'''
```

```python
'''Use | for OR
(condition1) | (condition2)'''
```

```python
'''Always Use Parentheses

Correct:

(df["sales"] > 1000)

Wrong:

df["sales"] > 1000

inside multiple conditions.'''
```

```python
#Example — OR Condition
df[(df["region"] == "North") | (df["region"] == "West")]
```

```python
'''PART 6 — Sorting Data

Sorting helps arrange data in order.'''
```

```python
#Sort by Sales (Ascending)
df.sort_values("sales")
```

```python
#Sort by Sales (Descending)
df.sort_values("sales", ascending=False)
```

```python
#Sort by Multiple Columns
df.sort_values(["region", "sales"])
```

```python
'''LEVEL 4 — Data Cleaning (VERY IMPORTANT)
Most real jobs involve dirty data.

Missing Values'''


df.isnull()
df.dropna()
df.fillna(0)
```

```python
#Remove Duplicates
df.drop_duplicates()
```

```python
#Rename Columns
df.rename(columns={"sales":"total_sales"})
```

```python
#Change Data Types
df["sales"] = df["sales"].astype(int)
```

```python
'''LEVEL 5 — Calculations & Analysis
Now you become a data analyst.

Learn:

Sum'''

df["sales"].sum()
```

```python
#Average
df["sales"].mean()
```

```python
#Maximum
df["sales"].max()
```

```python
#Group By
df.groupby("region")["sales"].sum()
```

```python
'''LEVEL 6 — Working With Excel Files

Very useful for business jobs.

Read Excel
df = pd.read_excel("sales.xlsx")
Export Excel
df.to_excel("output.xlsx", index=False)'''
```

```python
'''LEVEL 7 — Charts & Visualization

Using Matplotlib

Basic Chart


Learn:

bar chart
line chart
pie chart
histogram'''

import matplotlib.pyplot as plt

df["sales"].plot(kind="bar")

plt.show()
```
